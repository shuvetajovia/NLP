TOKENIZATION

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
sentences = ['I love you',
             "I'm love myself "
             'My day 2 progress ']
```

```python
token = Tokenizer(num_words=100 , oov_token= "<OOV>")
token.fit_on_texts(sentences)
wordindex = token.word_index
print(wordindex)
```

Sequences

```python
sequence = token.texts_to_sequences(sentences)
padded = pad_sequences(sequence)
postpadding = pad_sequences(sequence , padding="post")
print(sequence,"\n")
print(padded,"\n")
print(postpadding)
```

Embedding

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
vocab_size = len(wordindex) + 1
embedding_dim = 16
max_length = padded.shape[1]
```

```python
model = Sequential()
model.add(Embedding(vocab_size, embedding_dim, input_length=max_length))
model.add(LSTM(32))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```
